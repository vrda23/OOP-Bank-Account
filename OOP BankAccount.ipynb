```python

import logging
import sys
formatter = logging.Formatter('[%(asctime)s] %(message)s')
logger = logging.getLogger(__name__)

file_handler = logging.FileHandler("bankaccount.log")
stream_handler = logging.StreamHandler(sys.stdout)
class BankAccount(object):
  balance = 0
  def __init__(self, name):
    self.name = name

  def __repr__(self):
    return "Account name: {}, Balance: {}".format(self.name, self.balance)

  def show_balance(self):
    print("Balance: {}".format(self.balance))
    return

  def deposit(self, amount):
    logger.info("Depositing dollar amount to bank account")
    if amount <= 0:
      logger.error("Invalid value entered!")
      print("\nInvalid number entered. Quantity must be >0!")
      return
    else:
      print("You are depositing {}".format(amount))
      self.balance += amount
      self.show_balance()
    

  def withdraw(self, amount):
    if amount > self.balance:
      logger.error("Trying to withdraw above balance!")
      print("Invalid withdrawal!")
      return
    else:
      print("You are withdrawing {}".format(amount))
      self.balance -= amount
      self.show_balance()

my_account = BankAccount("Josip")
print(my_account)
my_account.show_balance()
my_account.deposit(2000)
my_account.deposit(-2000)
my_account.withdraw(1000)
print(my_account)

```
